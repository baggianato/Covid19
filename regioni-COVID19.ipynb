# COVID-19 regioni italiane - Raccolta e rappresentazione dati

Il seguente codice intende soltanto raccogliere e mostrare i dati raccolti dal Dipartimento di Protezione Civile. 

**Nessuna analisi viene effettuata.**

*fonte dati: https://github.com/pcm-dpc/COVID-19*

```python
import pandas as pd
import pylab
import numpy as np
```

```python
df_reg = pd.read_csv('./dati/dati-regioni/dpc-covid19-ita-regioni.csv', index_col="denominazione_regione")
```

```python
regione = 'Calabria'
dato = [
    'ricoverati_con_sintomi',
    'terapia_intensiva',
    'totale_ospedalizzati',
    'isolamento_domiciliare',
    'totale_attualmente_positivi',
    'nuovi_attualmente_positivi',
    'dimessi_guariti','deceduti',
    'totale_casi',
    'tamponi'
] 
dati_reg = df_reg.loc[regione]
```

```python
giorni=np.arange(len(dati_reg))
fig=pylab.figure(figsize=(20, 25), dpi= 80, facecolor='w', edgecolor='k')
fig.suptitle("Dati regione %s" % regione, fontsize=32, y=0.95)
i=1
while i<len(dato):
    for tipodato in dato:
        occo = dati_reg[tipodato].to_numpy()
        pylab.rc('font', size=15)
        pylab.subplot(5, 2, i)
        pylab.title('%s' % tipodato)
        if i==len(dato)-1 or i == len(dato):
            pylab.xlabel('Giorni dal 24/02/2020')
        pylab.ylabel('Occorrenze')
        pylab.grid(color='gray')
        #pylab.yscale('log')
        pylab.plot(giorni, occo, marker='d', linestyle='', color='b')
        i=i+1
```
